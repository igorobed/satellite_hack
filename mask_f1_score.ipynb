```python
pred_mask = cv2.imread(pred_mask_path)
pred_mask = cv2.cvtColor(pred_mask, cv2.COLOR_BGR2GRAY)
orig_mask = cv2.imread(orig_mask_path)
orig_mask = cv2.cvtColor(orig_mask, cv2.COLOR_BGR2GRAY)
```
